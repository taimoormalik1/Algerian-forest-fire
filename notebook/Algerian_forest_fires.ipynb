```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.info()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(labels=122,axis=0).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
##There is extra spaces in the columns
```

```python
##fix spaces
df.columns=df.columns.str.strip()
```

```python
df.head()
```

```python
##change the datatype as int
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
features=[features for features in df.columns if df[features].dtypes=='O'] 
```

```python
for i in features:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.to_csv('cleaned_algerian.csv',index=False)
```

```python
#Monthly fire analysis
dftemp=df.loc[df['Region']==1]
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.title('1ST REGION')
plt.show()
```
